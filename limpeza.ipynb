```python
#salarios
#remover NAS e substituir pelas medianas
dataset["Salario"].describe()
```

```python
#preenche os NAS com masculino, por ser o que mais ocorre
dataset["Genero"].fillna("Masculino", inplace = True)
dataset["Genero"].isnull().sum()
```
