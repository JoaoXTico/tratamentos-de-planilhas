```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.shape
```

```python
base.head()
```

```python
sbn.histplot(base.Volume, color='blue', kde=True, stat="density", bins=10).set(title="Volume")
```

```python
#carregar outra base de dados
base2 = pd.read_csv("chicken.csv")
base2.head()
```

```python
#criar dataframe agrupando o grupo "feed"
agrupar = base2.groupby(["feed"])["weight"].sum()
agrupar
```

```python
#filtrar
teste = base2.loc[base2['feed'] == 'horsebean']
teste
```

```python
#histograma levando em conta somente 'horsebean'
sbn.kdeplot(base2.loc[base2['feed'] == 'horsebean'].weight).set(title = 'horsebean')
```

```python
#histograma considerando somente o valor 'casein'
sbn.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde = True).set(title = 'casein')
```

```python
#linseed
sbn.histplot(base2.loc[base2['feed'] == 'linseed'].weight, kde = True).set(title = 'linseed')
```

```python
#meatmeal
sbn.histplot(base2.loc[base2['feed'] == 'meatmeal'].weight, kde = True).set(title = 'meat meal')
```

```python
#soybean
sbn.histplot(base2.loc[base2['feed'] == 'soybean'].weight, kde = True).set(title = 'soy bean')
```

```python
#sunflower
sbn.histplot(base2.loc[base2['feed'] == 'sunflower'].weight, kde = True).set(title = 'sunflower')
```

```python
#impresso em grfico 2x3 - 2 colunas e 3 linhas
plt.figure()
plt.subplot(3,2,1)
sbn.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde = True).set(title = 'casein')
plt.subplot(3,2,2)
sbn.histplot(base2.loc[base2['feed'] == 'linseed'].weight, kde = True).set(title = 'linseed')
plt.subplot(3,2,3)
sbn.histplot(base2.loc[base2['feed'] == 'meatmeal'].weight, kde = True).set(title = 'meat meal')
plt.subplot(3,2,4)
sbn.histplot(base2.loc[base2['feed'] == 'soybean'].weight, kde = True).set(title = 'soy bean')
plt.subplot(3,2,5)
sbn.histplot(base2.loc[base2['feed'] == 'sunflower'].weight, kde = True).set(title = 'sunflower')
plt.subplot(3,2,6)
sbn.histplot(base2.loc[base2['feed'] == 'horsebean'].weight, kde = True).set(title = 'horsebean')

#para no haver sobreposio
plt.tight_layout()
```
